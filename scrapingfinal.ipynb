BRA-MEX-COL

```python
def setup_driver():
    """Configura y retorna el WebDriver."""
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Ejecuta el navegador en segundo plano
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver

def get_club_names(driver, select_element_id):
    """Obtiene los nombres de los clubes desde el menú desplegable especificado."""
    driver.get('https://www.capology.com/')  # URL base o la página que contiene el menú desplegable
    time.sleep(5)  # Espera a que el contenido se cargue

    try:
        # Encuentra el menú desplegable por ID
        select_element = driver.find_element(By.ID, select_element_id)
        
        # Obtiene todos los elementos <option> dentro del menú desplegable
        options = select_element.find_elements(By.TAG_NAME, 'option')
        
        # Extrae los valores y textos de las opciones
        clubs = [(option.text, option.get_attribute('value')) for option in options if option.get_attribute('value')]

        return clubs

    except NoSuchElementException:
        return f"No se encontró el menú desplegable con ID '{select_element_id}'."
    except Exception as e:
        return f"Error: {e}"

def scrape_element(driver, url, element_id):
    """Raspa el texto del elemento especificado en la URL dada."""
    try:
        driver.get(url)
        time.sleep(5)  # Espera a que el contenido se cargue
        
        # Encuentra el elemento por ID usando Selenium
        element = driver.find_element(By.ID, element_id)
        value = element.text.strip()
        return value

    except NoSuchElementException:
        return f"Elemento con ID '{element_id}' no encontrado en la página."
    except Exception as e:
        return f"Error: {e}"

def main():
    driver = setup_driver()
    
    # Lista de IDs de menús desplegables a raspar
    select_element_ids = ['brasileiro-salaries-club', 'categoria-primera-a-salaries-club', 'liga-mx-salaries-club']
    element_ids = ['weekly-avg', 'annual-avg']
    
    for select_element_id in select_element_ids:
        clubs = get_club_names(driver, select_element_id)
        
        if isinstance(clubs, str):
            print(clubs)
            continue
        
        results = []  # Lista para almacenar los resultados
        for club_name, club_url_suffix in clubs:
            url = f"https://www.capology.com{club_url_suffix}"
            print(f"Scraping URL: {url}")
            
            for element_id in element_ids:
                value = scrape_element(driver, url, element_id)
                print(f"Valor del elemento con ID '{element_id}' para el club '{club_name}': {value}")
                
                # Guarda los resultados en la lista
                results.append({
                    'club_name': club_name,
                    'url': url,
                    'element_id': element_id,
                    'value': value
                })

        # Guardar los resultados en un archivo CSV separado para cada menú desplegable
        df = pd.DataFrame(results)
        df.to_csv(f'{select_element_id}_salaries.csv', index=False)
        print(f"Archivo '{select_element_id}_salaries.csv' guardado.")
    
    driver.quit()
    
if __name__ == '__main__':
    main()
```

```python
def setup_driver():
    """Configura y retorna el WebDriver."""
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Ejecuta el navegador en segundo plano
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver

def get_club_names(driver, select_element_id):
    """Obtiene los nombres de los clubes desde el menú desplegable especificado."""
    driver.get('https://www.capology.com/')  # URL base o la página que contiene el menú desplegable
    time.sleep(5)  # Espera a que el contenido se cargue

    try:
        # Encuentra el menú desplegable por ID
        select_element = driver.find_element(By.ID, select_element_id)
        
        # Obtiene todos los elementos <option> dentro del menú desplegable
        options = select_element.find_elements(By.TAG_NAME, 'option')
        
        # Extrae los valores y textos de las opciones
        clubs = [(option.text, option.get_attribute('value')) for option in options if option.get_attribute('value')]

        return clubs

    except NoSuchElementException:
        return f"No se encontró el menú desplegable con ID '{select_element_id}'."
    except Exception as e:
        return f"Error: {e}"

def scrape_element(driver, url, element_id):
    """Raspa el texto del elemento especificado en la URL dada."""
    try:
        driver.get(url)
        time.sleep(5)  # Espera a que el contenido se cargue
        
        # Encuentra el elemento por ID usando Selenium
        element = driver.find_element(By.ID, element_id)
        value = element.text.strip()
        return value

    except NoSuchElementException:
        return f"Elemento con ID '{element_id}' no encontrado en la página."
    except Exception as e:
        return f"Error: {e}"

def main():
    driver = setup_driver()
    
    # Lista de IDs de menús desplegables a raspar
    select_element_ids = ["austrian-bundesliga-salaries-club" , "first-division-a-salaries-club" , "superligaen-salaries-club" , "veikkausliiga-salaries-club"  ]
    element_ids = ['weekly-avg', 'annual-avg']
    
    for select_element_id in select_element_ids:
        clubs = get_club_names(driver, select_element_id)
        
        if isinstance(clubs, str):
            print(clubs)
            continue
        
        results = []  # Lista para almacenar los resultados
        for club_name, club_url_suffix in clubs:
            url = f"https://www.capology.com{club_url_suffix}"
            print(f"Scraping URL: {url}")
            
            for element_id in element_ids:
                value = scrape_element(driver, url, element_id)
                print(f"Valor del elemento con ID '{element_id}' para el club '{club_name}': {value}")
                
                # Guarda los resultados en la lista
                results.append({
                    'club_name': club_name,
                    'url': url,
                    'element_id': element_id,
                    'value': value
                })

        # Guardar los resultados en un archivo CSV separado para cada menú desplegable
        df = pd.DataFrame(results)
        df.to_csv(f'{select_element_id}_salaries.csv', index=False)
        print(f"Archivo '{select_element_id}_salaries.csv' guardado.")
    
    driver.quit()
    
if __name__ == '__main__':
    main()
```
